## 01. Gemini API

For a **Gemini API** case, you can manage your API key securely using a `.env` file, similar to the OpenAI setup. Here's how you can do it:

First, install the `python-dotenv` package to manage your environment variables:

```bash
pip3 install python-dotenv
```

Next, in the root directory of your project, create a `.env` file and store your Gemini API key like this:

```python
GEMINI_API_KEY="YOUR_API_KEY"
```

> Ensure that your `.env` file is excluded from version control by adding it to `.gitignore`:
>
> ```env
> .env
> ```

````python
# Import necessary libraries
import os
import google.generativeai as genai
from dotenv import load_dotenv

# Load the .env file
load_dotenv()

# Set the API key
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))
````

## 02. Google Gemini Chat Completion

The simplest way to generate text using the Gemini API is to provide the model with a single text-only input. Let's try it out using zero-shot approach!

````python
model = genai.GenerativeModel("gemini-1.5-flash")
response = model.generate_content("Write about blackhole and what happens when you approach event horizon of a blackhole.")

print(response.text)
````
